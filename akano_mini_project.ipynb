```python
'''to replace missing values, the distributuion, and description of the data will be considered
mean is usually used for a data that is well distributed and progressing predictable, if it is not 
skewed when visialised and the mean is close to the median value median is usually used 
if the standard deviation is too high in relation to the mean, if the scale of the data
is changing so much and if it looks skewed when visualised or not well distributed...
using these conditions, missing values for the numerical variables will be replaced as follows
'account_length'- median, 'number_vmail_messages'- median, 'total_day_minutes'- median,
'total_day_calls'- mean, 'total_day_charge'- mean, 'total_eve_minutes'- median, 
'total_eve_calls'- mean, 'total_eve_charge'- mean, 'total_night_minutes'-mean,
'total_night_charge'- mean, 'total_intl_minutes'- mean, 'total_intl_calls'- median,
'total_intl_charge- mean', 'number_customer_service_calls'- median'''
```

```python
#replace missing values
data['account_length'] = data['account_length'].fillna(data['account_length'].median())
data['number_vmail_messages'] = data['number_vmail_messages'].fillna(data['number_vmail_messages'].median())
data['total_day_minutes'] = data['total_day_minutes'].fillna(data['total_day_minutes'].median())
data['total_day_calls'] = data['total_day_calls'].fillna(data['total_day_calls'].mean())
data['total_day_charge'] = data['total_day_charge'].fillna(data['total_day_charge'].mean())
data['total_eve_minutes'] = data['total_eve_minutes'].fillna(data['total_eve_minutes'].median())
data['total_eve_calls'] = data['total_eve_calls'].fillna(data['total_eve_calls'].mean())
data['total_eve_charge'] = data['total_eve_charge'].fillna(data['total_eve_charge'].mean())
data['total_night_minutes'] = data['total_night_minutes'].fillna(data['total_night_minutes'].mean())
data['total_night_charge'] = data['total_night_charge'].fillna(data['total_night_charge'].mean())
data['total_intl_minutes'] = data['total_intl_minutes'].fillna(data['total_intl_minutes'].mean())
data['total_intl_calls'] = data['total_intl_calls'].fillna(data['total_intl_calls'].median())
data['total_intl_charge'] = data['total_intl_charge'].fillna(data['total_intl_charge'].mean())
data['number_customer_service_calls'] = data['number_customer_service_calls'].fillna(data['number_customer_service_calls'].median())
```

# Machine Learning Models

# Train and Test Sets

```python
# Import Libraries and packeges
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# Import data and split into X and Y variablesabs
dataframe = data
array = dataframe.values
X = array[:,0:19]
Y = array[:,19]

# Specify split ration for Test Data
test_size = 0.1
seed = 7
# Call train_test_split() and pass X, Y and Split ratio as parametrs
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,random_state=seed)

#Call algorithm construtor and pass to Model varible
model = LogisticRegression()# At this point model is illitrate

# Traing model by calling model.fit() and pass training data(x& Y)
model.fit(X_train, Y_train) # At this point, model has learnt a lot!

# Assess model performance by calling model.score() and pass test data (X,Y)
result = model.score(X_test, Y_test)# Model nolonger learning. We are testing it using our test data to see its performance
#print("Accuracy: {}".format(result*100.0))

print(result)
```

# K-fold Cross Validation

```python
# Evaluate using Cross Validation
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import statistics
dataframe = data
array = dataframe.values
X = array[:,0:19]
Y = array[:,19]

num_folds = 7
seed = 7
kfold = KFold(n_splits=num_folds)#, random_state=seed)
model = LogisticRegression(max_iter=1000)
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: {}.3f%%, {}.3f%%".format(results.mean()*100.0, results.std()*100.0))

print(results)
print(statistics.mean(results))
print(statistics.stdev(results))
```

# Leave One Out Cross Validation

```python
# Evaluate using Leave One Out Cross Validation
from pandas import read_csv
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import statistics
dataframe = data
array = dataframe.values
X = array[:,0:19]
Y = array[:,19]


array = dataframe.values
X = array[:,0:19]
Y = array[:,19]
num_folds = 10
loocv = LeaveOneOut()
model = LogisticRegression(max_iter=1000)
results = cross_val_score(model, X, Y, cv=loocv)
print("Accuracy: {} ({})".format(results.mean()*100.0, results.std()))#*100.0)) 
#print(statis)
```

# Shuffle Split Cross Validation

```python
# Evaluate using Shuffle Split Cross Validation
from pandas import read_csv
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import statistics
dataframe = data
array = dataframe.values
X = array[:,0:19]
Y = array[:,19]

n_splits = 10
test_size = 0.33
 
seed = 7
kfold = ShuffleSplit(n_splits=n_splits, test_size=test_size)#, random_state=seed)
model = LogisticRegression(max_iter=1000)
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: {} ({})".format(results.mean()*100.0, results.std()*100.0))
```
